```python
# Checking for missing data and NULL values in the dataframe

total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data
```

```python
explicit    = df[df['explicit'] == 1]
not_explicit = df[df['explicit'] == 0]
```

```python
print("Total =", len(df))

print("Explicit Songs =", len(explicit))
print("Percentage Explicit =", 1.*len(explicit)/len(df)*100.0, "%")
 
print("Not Explicit Songs =", len(not_explicit))
print("Percentage not Explicit =", 1.*len(not_explicit)/len(df)*100.0, "%")
```
